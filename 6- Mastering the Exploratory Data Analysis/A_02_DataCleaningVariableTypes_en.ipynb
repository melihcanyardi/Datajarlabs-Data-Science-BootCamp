**(2)** Using the `movie` data, we have studied in this lesson. Complete the following tasks:
    * Select a variable from the `movie` dataset and discuss the variable type of this variable
    * Categorize a variable of your choice and assign a name to it
    * Count the number of rows for each category

```python
import pandas as pd
import numpy as np

movie = pd.read_csv("https://djl-lms-assets.s3.eu-central-1.amazonaws.com/datasets/movie.csv")
movie.head()
```

```python
print("Duration Variable Datatype:", movie['duration'].dtype, "\n")
display(movie['duration'].head())

print("\nSummary Statistics:\n", movie['duration'].describe(), sep='')
```

"Duration" variable has a continuous variable.

```python
def duration_categorizer(duration):
  if pd.isnull(duration):
    return np.nan
  elif duration < 40:
    return 'short'
  elif duration < 150:
    return 'medium'
  else:
    return 'long'

movie['duration_category'] = movie['duration'].apply(duration_categorizer)
```

```python
print("Number of rows of each category:\n")
movie['duration_category'].value_counts()
```

**(3)** Is it possible to convert categorical variable into numerical one? If so, how can you make that?

Yes, just as in the previous question, one can create a numeric value for each of the categories. Or it can be easily done with LabelEncoder in Scikit-Learn's preprocessing module.

```python
def duration_numeric(duration):
  if pd.isnull(duration):
    return np.nan
  elif duration == 'short':
    return 0
  elif duration == 'medium':
    return 1
  else:
    return 2

movie['duration_category_numeric'] = movie['duration_category'].apply(duration_numeric)

movie.head()
```
