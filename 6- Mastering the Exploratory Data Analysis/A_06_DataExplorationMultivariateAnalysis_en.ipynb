```python
df = pd.read_csv('StudentsPerformance.csv')

df.head()
```

```python
df['total score'] = (df['math score'] + df['reading score'] + df['writing score']) / 3
```

**(1)** Does the performance in the exams vary according to gender, origin and education level of parents?

```python
print("Gender:")
display(df.groupby('gender')[['math score', 'reading score', 'writing score', 'total score']].mean().T)

print("\nOrigin:")
display(df.groupby('race/ethnicity')[['math score', 'reading score', 'writing score', 'total score']].mean().T)

print("\nEducation level of parents:")
display(df.groupby('parental level of education')[['math score', 'reading score', 'writing score', 'total score']].mean().T)
```

```python
plt.figure(figsize=(24, 20))

plt.subplot(3, 4, 1)
sns.barplot(x='gender', y='math score', data=df)
plt.title('Gender & Math Score')

plt.subplot(3, 4, 2)
sns.barplot(x='gender', y='reading score', data=df)
plt.title('Gender & Reading Score')

plt.subplot(3, 4, 3)
sns.barplot(x='gender', y='writing score', data=df)
plt.title('Gender & Writing Score')

plt.subplot(3, 4, 4)
sns.barplot(x='gender', y='total score', data=df)
plt.title('Gender & Total Score')

plt.subplot(3, 4, 5)
sns.barplot(x='race/ethnicity', y='math score', data=df)
plt.title('Origin & Math Score')

plt.subplot(3, 4, 6)
sns.barplot(x='race/ethnicity', y='reading score', data=df)
plt.title('Origin & Reading Score')

plt.subplot(3, 4, 7)
sns.barplot(x='race/ethnicity', y='writing score', data=df)
plt.title('Origin & Writing Score')

plt.subplot(3, 4, 8)
sns.barplot(x='race/ethnicity', y='total score', data=df)
plt.title('Origin & Total Score')

plt.subplot(3, 4, 9)
sns.barplot(x='parental level of education', y='math score', data=df)
plt.title('Education Level of Parents & Math Score')
plt.xticks(rotation=90)

plt.subplot(3, 4, 10)
sns.barplot(x='parental level of education', y='reading score', data=df)
plt.title('Education Level of Parents & Reading Score')
plt.xticks(rotation=90)

plt.subplot(3, 4, 11)
sns.barplot(x='parental level of education', y='writing score', data=df)
plt.title('Education Level of Parents & Writing Score')
plt.xticks(rotation=90)

plt.subplot(3, 4, 12)
sns.barplot(x='parental level of education', y='total score', data=df)
plt.title('Education Level of Parents & Total Score')
plt.xticks(rotation=90)

plt.show()
```
