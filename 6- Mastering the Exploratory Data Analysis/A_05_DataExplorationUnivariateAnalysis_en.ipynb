**(5)** You may have noticed that there are many missing values in the mathematics and reading notes. Complete the missing values with mean, median, and interpolation. Compare the impact of these techniques on the distribution of grades.

```python
df = pd.read_csv('states_all.csv')

df.isna().sum()
```

```python
plt.figure(figsize=(25, 20))

plt.subplot(3, 4, 1)
plt.hist(df['AVG_MATH_4_SCORE'].fillna(df['AVG_MATH_4_SCORE'].mean()))
plt.title('AVG_MATH_4_SCORE (Mean)', fontsize=15)

plt.subplot(3, 4, 2)
plt.hist(df['AVG_MATH_8_SCORE'].fillna(df['AVG_MATH_8_SCORE'].mean()))
plt.title('AVG_MATH_8_SCORE (Mean)', fontsize=15)

plt.subplot(3, 4, 3)
plt.hist(df['AVG_READING_4_SCORE'].fillna(df['AVG_READING_4_SCORE'].mean()))
plt.title('AVG_READING_4_SCORE (Mean)', fontsize=15)

plt.subplot(3, 4, 4)
plt.hist(df['AVG_READING_8_SCORE'].fillna(df['AVG_READING_8_SCORE'].mean()))
plt.title('AVG_READING_8_SCORE (Mean)', fontsize=15)

plt.subplot(3, 4, 5)
plt.hist(df['AVG_MATH_4_SCORE'].fillna(df['AVG_MATH_4_SCORE'].median()))
plt.title('AVG_MATH_4_SCORE (Median)', fontsize=15)

plt.subplot(3, 4, 6)
plt.hist(df['AVG_MATH_8_SCORE'].fillna(df['AVG_MATH_8_SCORE'].median()))
plt.title('AVG_MATH_8_SCORE (Median)', fontsize=15)

plt.subplot(3, 4, 7)
plt.hist(df['AVG_READING_4_SCORE'].fillna(df['AVG_READING_4_SCORE'].median()))
plt.title('AVG_READING_4_SCORE (Median)', fontsize=15)

plt.subplot(3, 4, 8)
plt.hist(df['AVG_READING_8_SCORE'].fillna(df['AVG_READING_8_SCORE'].median()))
plt.title('AVG_READING_8_SCORE (Median)', fontsize=15)

plt.subplot(3, 4, 9)
plt.hist(df['AVG_MATH_4_SCORE'].interpolate())
plt.title('AVG_MATH_4_SCORE (Interpolation)', fontsize=15)

plt.subplot(3, 4, 10)
plt.hist(df['AVG_MATH_8_SCORE'].interpolate())
plt.title('AVG_MATH_8_SCORE (Interpolation)', fontsize=15)

plt.subplot(3, 4, 11)
plt.hist(df['AVG_READING_4_SCORE'].interpolate())
plt.title('AVG_READING_4_SCORE (Interpolation)', fontsize=15)

plt.subplot(3, 4, 12)
plt.hist(df['AVG_READING_8_SCORE'].interpolate())
plt.title('AVG_READING_8_SCORE (Interpolation)', fontsize=15)

plt.show()
```
