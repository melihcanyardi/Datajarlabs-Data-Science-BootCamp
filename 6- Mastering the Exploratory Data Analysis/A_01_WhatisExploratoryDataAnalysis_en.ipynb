Because most machine learning models and even some plots (e.g. boxplot) does not work when there are missing values.

**(4)** Do you think that outliers have an impact on a dataset? If so, how would you explain this impact?

They definitely have an impact on the dataset. Firstly, they have an important impact on the standard deviation, which will make the distribution of the variable change a lot. Also, having outlier in the dataset will make some machine learning models (e.g., Linear Regression) estimate worse than when the outliers are removed.
