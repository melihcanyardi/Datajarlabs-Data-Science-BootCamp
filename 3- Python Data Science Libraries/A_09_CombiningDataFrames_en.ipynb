````python
# Creating first DataFrame
df1_cols = ['id', 'age', 'sex', 'beats_per_sec', 'cholesterol']
df1_data = np.array([
    [1001,63,'Male',145,233],
    [1001,37,'Male',130,250],
    [1002,41,'Female',130,204],
    [1002,56,'Male',120,236],
    [1003,57,'Female',120,354],
    [1004,57,'Male',140,192],
    [1005,56,'Female',140,294],
    [1006,44,'Male',120,263],
    [1007,52,'Male',172,199],
    [1007,57,'Male',150,168],
    [1008,54,'Male',140,239],
    [1008,48,'Female',150,264]
])

df_1 = pd.DataFrame(data = df1_data, columns = df1_cols)
df_1['id'] = df_1['id'].astype(int)
df_1['age'] = df_1['age'].astype(int)
df_1['beats_per_sec'] = df_1['beats_per_sec'].astype(int)
df_1['cholesterol'] = df_1['cholesterol'].astype(int)
````

````python
# Creating second DataFrame
df2_cols = ['id', 'target']
df2_data = np.array([
    [1001,0],
    [1002,1],
    [1004,0],
    [1005,1],
    [1007,0],
    [1008,1]
])

df_2 = pd.DataFrame(data = df2_data, columns = df2_cols)
df_2['id'] = df_2['id'].astype(int)
df_2['target'] = df_2['target'].astype(int)
````

````python
# Merging two DataFrames
inner_join = pd.merge(df_1, df_2, on='id')
print('Inner join:')
display(inner_join)

outer_join = pd.merge(df_1, df_2, on='id', how='outer')
print('Outer join:')
display(outer_join)

left_join = pd.merge(df_1, df_2, how='left')
print('Left join:')
display(left_join)

right_join = pd.merge(df_1, df_2, how='right')
print('Right join:')
display(right_join)
````
