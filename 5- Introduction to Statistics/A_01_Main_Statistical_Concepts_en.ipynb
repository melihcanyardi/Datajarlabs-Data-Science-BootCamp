```python
import numpy as np
import pandas as pd
import statistics
```

- Analyze the central tendency and distribution measurements with the mathematical formula given above and the built-in codes available in python through 3 different data that you will generate using the numpy library !!!

```python
random_1 = np.random.normal(30, 15, 100)
random_2 = np.random.normal(100, 20, 1001)
random_3 = np.random.normal(250, 100, 10000)
```

## Mean

```python
mean_1 = sum(random_1) / len(random_1)
mean_2 = sum(random_2) / len(random_2)
mean_3 = sum(random_3) / len(random_3)

print("Mean 1: {}".format(mean_1))
print("Mean 2: {}".format(mean_2))
print("Mean 3: {}\n".format(mean_3))

#print("Mean 1: {}".format(np.mean(random_1)))
#print("Mean 2: {}".format(np.mean(random_2)))
#print("Mean 3: {}".format(np.mean(random_3)))
```

## Median

```python
def median(data):
  if len(data) % 2 == 1:
    median = sorted(data)[int(len(data)/2)]
  else:
    median = (sorted(data)[int(len(data)/2)] + sorted(data)[int(len(data)/2)-1]) / 2
  
  return median
```

```python
print("Median 1: {}".format(median(random_1)))
print("Median 1: {}".format(np.median(random_1)))
print("Median 1: {}\n".format(statistics.median(random_1)))

print("Median 2: {}".format(median(random_2)))
print("Median 2: {}".format(np.median(random_2)))
print("Median 2: {}\n".format(statistics.median(random_2)))

print("Median 3: {}".format(median(random_3)))
print("Median 3: {}".format(np.median(random_3)))
print("Median 3: {}\n".format(statistics.median(random_3)))
```

## Mode

```python
np.random.seed(42)

random_int_1 = np.random.randint(10, 20, 25)
random_int_2 = np.random.randint(30, 60, 100)
random_int_3 = np.random.randint(100, 500, 10000)
```

```python
print("Mode 1: {}".format(statistics.mode(random_int_1)))
print("Mode 2: {}".format(statistics.mode(random_int_2)))
print("Mode 3: {}".format(statistics.mode(random_int_3)))
```

```python
(values, counts) = np.unique(random_int_1, return_counts=True)
ind = np.argmax(counts)
print("Mode 1: {}".format(values[ind]))

(values, counts) = np.unique(random_int_2, return_counts=True)
ind = np.argmax(counts)
print("Mode 2: {}".format(values[ind]))

(values, counts) = np.unique(random_int_3, return_counts=True)
ind = np.argmax(counts)
print("Mode 3: {}".format(values[ind]))
```

## Variance

```python
print("Variance 1: {}".format(np.var(random_1, ddof=1)))
print("Variance 2: {}".format(np.var(random_2, ddof=1)))
print("Variance 3: {}\n".format(np.var(random_3, ddof=1)))

print("Variance 1: {}".format(random_1.var(ddof=1)))
print("Variance 2: {}".format(random_2.var(ddof=1)))
print("Variance 3: {}".format(random_3.var(ddof=1)))
```

## Standard Deviation

```python
print("Standard Deviation 1: {}".format(np.std(random_1)))
print("Standard Deviation 2: {}".format(np.std(random_2)))
print("Standard Deviation 3: {}".format(np.std(random_3)))
```

## Standard Error

```python
print("Standard Error 1: {}".format(np.std(random_1) / np.sqrt(len(random_1))))
print("Standard Error 2: {}".format(np.std(random_2) / np.sqrt(len(random_2))))
print("Standard Error 3: {}".format(np.std(random_3) / np.sqrt(len(random_3))))
```
